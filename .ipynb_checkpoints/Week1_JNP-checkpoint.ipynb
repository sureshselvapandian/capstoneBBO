```python
from pathlib import Path

import numpy as np

ROOT = Path().resolve()
INITIAL_DATA_DIR = ROOT / "data" / "initial_data"
WEEK1_DIR = ROOT / "data" / "week_1"


def propose_query(inputs: np.ndarray, outputs: np.ndarray, rng: np.random.Generator) -> np.ndarray:
    best_index = int(np.argmax(outputs))
    base_point = inputs[best_index].astype(float).copy()

    scale = 0.05
    for _ in range(200):
        candidate = np.clip(base_point + rng.normal(0.0, scale, size=base_point.shape), 0.0, 1.0)
        candidate = np.round(candidate, 6)
        if not np.any(np.all(np.isclose(inputs, candidate, atol=1e-8), axis=1)):
            return candidate
        scale *= 1.03

    return np.round(base_point, 6)


def generate_week1_inputs(seed: int = 42) -> Path:
    rng = np.random.default_rng(seed)
    WEEK1_DIR.mkdir(parents=True, exist_ok=True)

    summary_lines = ["function,dimension,week1_input_file"]

    for function_id in range(1, 9):
        function_dir = INITIAL_DATA_DIR / f"function_{function_id}"
        inputs = np.load(function_dir / "initial_inputs.npy")
        outputs = np.load(function_dir / "initial_outputs.npy")

        query = propose_query(inputs, outputs, rng).reshape(1, -1)

        out_dir = WEEK1_DIR / f"function_{function_id}"
        out_dir.mkdir(parents=True, exist_ok=True)
        out_file = out_dir / "week_1_input.npy"
        np.save(out_file, query)

        summary_lines.append(f"{function_id},{query.shape[1]},{out_file.as_posix()}")

    index_file = WEEK1_DIR / "week_1_queries_index.csv"
    index_file.write_text("\n".join(summary_lines), encoding="utf-8")
    return index_file
```

```python
index_file = generate_week1_inputs(seed=42)
print(f"Saved index: {index_file}")

for function_id in range(1, 9):
    query_file = WEEK1_DIR / f"function_{function_id}" / "week_1_input.npy"
    query = np.load(query_file)
    print(f"Function {function_id} -> shape {query.shape} -> query {query.tolist()}")
```
