```python
import numpy as np
import sys 
import os
import ast
import re
sys.path.append(os.path.abspath(".."))
from black_box_optimizer import optimise_black_box


PROJECT_ROOT = os.path.abspath("..")
DATA_DIR = os.path.join(PROJECT_ROOT, "queries")

results = {}

for FUNCTION in range(1, 9):
    print(f"\n==============================")
    print(f" Running optimisation for FUNCTION {FUNCTION}")
    print(f"==============================")

    # Load initial data
    X0_path = os.path.join(DATA_DIR, f"week1/function_{FUNCTION}_inputs.txt")
    y0_path = os.path.join(DATA_DIR, f"week1/function_{FUNCTION}_outputs.txt")
    with open(X0_path, "r") as f:
        txt = f.read()

    print(txt)
    # Convert string → Python list of arrays 
    clean_txt = re.sub(r'array\(', '(', txt)
    python_list = ast.literal_eval(clean_txt)
    # Convert to a proper NumPy array 
    X0 = np.array(python_list)

    with open(y0_path, "r") as f:
        raw = f.read()

    print(raw)
    clean = re.sub(r'np\.float64\((.*?)\)', r'\1', raw)
    python_list = ast.literal_eval(clean)
    y0 = np.array(python_list, dtype=float)

    print("Initial shapes:", X0.shape, y0.shape)

    # Run optimisation
    all_X, all_y = optimise_black_box(
        initial_inputs=X0,
        initial_outputs=y0,
        max_queries=1,   # adjust as needed
        seed=42,
    )

    # Best result
    best_idx = np.argmax(all_y)
    best_x = all_X[best_idx]
    best_y = all_y[best_idx]

    print("Best x:", best_x)
    print("Best y:", best_y)

    # Store results
    results[FUNCTION] = {
        "best_x": best_x,
        "best_y": best_y,
        "all_X": all_X,
        "all_y": all_y,
    }

print("\nAll functions completed.")
```
