```python
import numpy as np
import sys 
import os
sys.path.append(os.path.abspath(".."))
from black_box_optimizer import optimise_black_box


PROJECT_ROOT = os.path.abspath("..")
#DATA_DIR = os.path.join(PROJECT_ROOT, "queries")

results = {}

for FUNCTION in range(1, 2):
    print(f"\n==============================")
    print(f" Running optimisation for FUNCTION {FUNCTION}")
    print(f"==============================")

    DATA_DIR = os.path.join(PROJECT_ROOT, "data")
    X0_path = os.path.join(DATA_DIR, f"initial_data/function_{FUNCTION}/initial_inputs.npy")
    y0_path = os.path.join(DATA_DIR, f"initial_data/function_{FUNCTION}/initial_outputs.npy")

    initial_X0 = np.load(X0_path)
    initial_y0 = np.load(y0_path)

    DATA_DIR = os.path.join(PROJECT_ROOT, "queries")
    # Load initial data
    X0_path = os.path.join(DATA_DIR, f"week1/input_function_{FUNCTION}.npy")
    y0_path = os.path.join(DATA_DIR, f"week1/output_function_{FUNCTION}.npy")

    next_X0 = np.load(X0_path)

    next_X0 = next_X0.reshape(1, -1)
    next_y0 = np.load(y0_path)

    X0 = initial_X0 # np.vstack((initial_X0, next_X0))
    y0 = np.concatenate((initial_y0, next_y0))

    # Print the raw arrays 
    print("\nX0:") 
    print(X0) 
    print("\ny0:") 
    print(y0)

    print("Initial shapes:", X0.shape, y0.shape)

    # Run optimisation
    all_X, all_y = optimise_black_box(
        initial_inputs=X0,
        initial_outputs=y0,
        max_queries=3,   # adjust as needed
        seed=42,
    )

    # Best result
    best_idx = np.argmax(all_y)
    best_x = all_X[best_idx]
    best_y = all_y[best_idx]

    print("Best x:", best_x)
    print("Best y:", best_y)

    # Store results
    results[FUNCTION] = {
        "best_x": best_x,
        "best_y": best_y,
        "all_X": all_X,
        "all_y": all_y,
    }

print("\nAll functions completed.")
```
