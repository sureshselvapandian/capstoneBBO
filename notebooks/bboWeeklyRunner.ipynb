# BBO Weekly Runner

> This notebook runs one optimisation step for each of the 8 black-box functions for a selected week.

## What this notebook does
- Reads `function_k_inputs.txt` and `function_k_outputs.txt` from `queries/<week>/`
- Converts text-based arrays into NumPy arrays
- Calls `optimise_black_box(...)` for each function
- Collects and stores best/complete results in `results`

> Update `WEEK` before running the notebook.

```python
import ast
import os
import re
import sys

import numpy as np

# Allow importing modules from the repository root
sys.path.append(os.path.abspath(".."))
from black_box_optimizer import optimise_black_box

# -----------------------------------------------------------------------------
# Configuration
# -----------------------------------------------------------------------------
PROJECT_ROOT = os.path.abspath("..")
DATA_DIR = os.path.join(PROJECT_ROOT, "queries")

# Pick the week folder to run against (e.g., week1, week2, ... week13)
#WEEK = "week1"
#WEEK = "week2"
#WEEK = "week3"
#WEEK = "week4"
#WEEK = "week5"
#WEEK = "week6"
#WEEK = "week7"
#WEEK = "week8"
#WEEK = "week9"
#WEEK = "week10"
#WEEK = "week11"
#WEEK = "week12"
WEEK = "week13"

# Container for per-function outputs
results = {}


def load_inputs_txt(file_path: str) -> np.ndarray:
    """Load input vectors from text in format like [array([...]), array([...]), ...]."""
    with open(file_path, "r", encoding="utf-8") as file:
        text = file.read()

    # Convert "array(...)" into plain tuples/lists for safe literal parsing
    cleaned_text = re.sub(r"array\(", "(", text)
    parsed = ast.literal_eval(cleaned_text)
    return np.array(parsed, dtype=float)


def load_outputs_txt(file_path: str) -> np.ndarray:
    """Load scalar outputs from text in format like [np.float64(...), ...]."""
    with open(file_path, "r", encoding="utf-8") as file:
        text = file.read()

    # Convert np.float64(x) -> x before parsing
    cleaned_text = re.sub(r"np\.float64\((.*?)\)", r"\1", text)
    parsed = ast.literal_eval(cleaned_text)
    return np.array(parsed, dtype=float)


for function_id in range(1, 9):
    print("\n==============================")
    print(f"Running optimisation for FUNCTION {function_id}")
    print("==============================")

    # Build paths for the selected week
    x_path = os.path.join(DATA_DIR, f"{WEEK}/function_{function_id}_inputs.txt")
    y_path = os.path.join(DATA_DIR, f"{WEEK}/function_{function_id}_outputs.txt")

    # Load historical data for this function
    x_initial = load_inputs_txt(x_path)
    y_initial = load_outputs_txt(y_path)
    print("Initial shapes:", x_initial.shape, y_initial.shape)

    # Run one optimisation proposal step
    all_x, all_y = optimise_black_box(
        initial_inputs=x_initial,
        initial_outputs=y_initial,
        max_queries=1,
        seed=42,
    )

    # Compute current best observed point/value
    best_idx = int(np.argmax(all_y))
    best_x = all_x[best_idx]
    best_y = float(all_y[best_idx])

    print("Best x:", best_x)
    print("Best y:", best_y)

    # Persist function-level results for later analysis
    results[function_id] = {
        "best_x": best_x,
        "best_y": best_y,
        "all_X": all_x,
        "all_y": all_y,
    }

print("\nAll functions completed.")
```
